```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data_directory = "./data/"
ratings_df = pd.read_csv(data_directory+'u.data', delimiter='\t' )
users_df = pd.read_csv(data_directory+'u.user', delimiter='|')
genre_df = pd.read_csv(data_directory+'u.genre', delimiter='|')
films_df = pd.read_csv(data_directory+'u.item', delimiter='|', encoding='ISO-8859-1')
occupation_df = pd.read_csv(data_directory+'u.occupation')
films_df
```

# remove the genre columns 
and insert a new column called genre that contains the name of the  columns when 1

```python
# Extracting genre names where value is 1 and creating a 'Genre' column
genre_columns = ['unknown', 'Action', 'Adventure', 'Animation', "Children's",
                 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir',
                 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War',
                 'Western']

# Function to get the first genre from columns where value is 1
def get_first_genre(row):
    for column in genre_columns:
        if row[column]:
            return column
    return 'Unknown'  # Default genre if no genre is found

# Create 'Genre' column by applying the function row-wise
films_df['Genre'] = films_df.apply(get_first_genre, axis=1)

# Drop the original genre columns
films_df.drop(columns=genre_columns, inplace=True)
films_df
```

```python

# Merging the DataFrames
merged_df = pd.merge(ratings_df, users_df, left_on='user_id', right_on='user id', how='inner')
# Merging the DataFrames
merged_df = pd.merge(ratings_df, users_df, left_on='user_id', right_on='user id', how='inner')

# Get the total count of ratings for each occupation
occupation_counts = merged_df.groupby('occupation')['rating'].count()

# Select top N professions with the highest counts
top_n = 5  # Replace '5' with the desired number of top professions
top_occupations = occupation_counts.nlargest(top_n).index

# Filter the merged DataFrame for the top occupations
filtered_df = merged_df[merged_df['occupation'].isin(top_occupations)]

# Calculate counts for each rating within each occupation
occupation_rating_counts = filtered_df.groupby(['rating', 'occupation']).size().unstack(fill_value=0)
# Plotting separate violin plots for each rating category
plt.figure(figsize=(12, 6))
sns.violinplot(x='occupation', hue='rating', data=filtered_df, palette='rainbow', split=True)
plt.title(f'Distribution of Ratings for Top {top_n} Occupations')
plt.xlabel('Occupation')
plt.ylabel('Rating')
plt.legend(title='Rating')
plt.xticks(rotation=45)
plt.show()
```
