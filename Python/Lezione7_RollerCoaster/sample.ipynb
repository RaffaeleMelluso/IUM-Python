```python
filtered_df=df.loc[~df['speed1_value'].isna()]
filtered_df[['coaster_name','speed1_value']]
```

```python
mean=filtered_df['speed1_value'].mean()
print("La velocità media è", mean)
```

```python
axis = df['speed1_value'].value_counts().plot(kind='bar', title="max_speed")
```

```python
# Calcola la media della colonna 'speed1_value'
mean_speed = filtered_df['speed1_value'].mean()

# Sostituisci i valori NaN nella colonna 'speed1_value' con la media
df.loc[df['speed1_value'].isna(), 'speed1_value'] = mean_speed

# Conta il numero di montagne russe con velocità media o sostituita
count = df.loc[df['speed1_value'] == mean_speed].count()
print("Il numero di montagne russe con velocità media o sostituita è", count['speed1_value'])
```
